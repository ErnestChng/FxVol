```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/ErnestChng/FxVol/main/data/df.csv")
df.ffill(axis = 0, inplace = True) 
df.index = pd.to_datetime(df.Date)
df.drop(columns=['Date'], inplace = True)

df.head()
```

```
print(df[df.isna().any(axis=1)])
# 0 - 13 contain NaN values. Remove them

df = df.loc['2013-02-02':]
```

```
ind = ["ATR", "OBV", "RSI", "BB_up", "BB_dn"]
col_x = [c for c in df.columns.values for i in ind if i in c]
col_y = df.drop(col_x, axis = 1).columns.values

df_y = df[col_y].pct_change().rolling(21).std().iloc[21:]
df_x = df[col_x].iloc[21:]

df = pd.merge(df_x, df_y,left_index=True, right_index=True)
```

```
df.sort_index(axis=1, inplace=True)
df.columns
```
